<a id=add><a/>
### Adding Row/Column

```python
# Dictionary
num_strs = ['one','two','three','four']
dictt = {'Numbers':[1,2,3,4], 'Squares':[1,4,9,16], 'Cubes':[1,16,27,64]}
df = pd.DataFrame(dictt, index=num_strs)
print(df)

print('\nAdd new columns:')
df['Negatives'] = df.Numbers * -1
df['Product'] = df.Numbers * df.Negatives
print(df)

print('\nAdd a new row (with index name):')
df.loc['five'] = [5, 25, 125, -5, -25]
print(df)

print('\nAdd a new row (with default viz. index position):')
df.loc[df.shape[0]] = [6, 36, 216, -6, -36]
print(df)

print('\nAdd a new row using PD Series:')
ser = pd.Series([7, 49, 343, -7, -49], name='seven', index=df.columns)
df = df.append(ser)     # Returns a new DataFrame, does not alter the existing DF
print(df)

print('\nAdd a new row using PD DataFrame:')
df2 = pd.DataFrame([[8, 64, 512, -8, -64]], columns=df.columns, index=['eight'])
df = df.append(df2)     # Returns a new DataFrame, does not alter the existing DF
print(df)

print('\nAdd new columns using PD Series:')
ser = pd.Series([1,1,1,1,1,1,1,1], name='^0')
df['^0'] = ser
print(df)
```

```python
market_df = pd.read_csv('http://pythonhow.com/supermarkets.csv')

print('\nStandard array slicing works only on rows:')
print(market_df[0:4])

# iloc[] - gets rows (or columns) at particular positions in the index (so it only takes integers).
print('\nUse iloc[] for 2D index position slicing:')
print(market_df.iloc[0:4, 0:4])

# loc[] - gets rows (or columns) with particular labels from the index.
print('\nUse loc[] for 2D index label slicing:')
print(market_df.loc[0:4, 'ID':'State']) # Here the row index pos and and label are the same, hence 0:4 can be used
```

<a id=CONCAT><a/>
### Concatenate

```python
emps = {'emp_id':[11001,11002,11003,11004,11005], 'emp_name':['Bob','Mary','Peter', 'Linda','Sam'], 'subject':['Physics','Biology','English','Chemistry','Philosophy']}
depts = {'dept_id':[55001,55002,55003,55004,55005], 'dept_name':['Science', 'Science', 'Science', 'Languages', 'Math'], 'subject':['Physics','Chemistry','Biology','English', 'Algebra'], 'difficulty':[3,3,3,1,3]}

emp_df = pd.DataFrame(emps)
dept_df = pd.DataFrame(depts)

print(emp_df)
print(dept_df)

row_concat = pd.concat([emp_df, dept_df])
col_concat = pd.concat([emp_df, dept_df], axis=1)

print('\nRow concat - default or axis=0')
print(row_concat)
print('\nColumn concat - axis=1')
print(col_concat)
```

<a id=APPEND><a/>
### Append
Equivalent to default `pd.concat()` (on `axis=0` which is the default)

**Note:**
If the DFs have common indices, the indices in the new DF will have repeating values.
* Ignore index
   * `df1.append(df2, ignore_index=True)`
* Reset index
   * `df1.append(df2)`
   * `df1.reset_index()`

```python
emps = {'emp_id':[11001,11002,11003,11004,11005], 'emp_name':['Bob','Mary','Peter', 'Linda','Sam'], 'subject':['Physics','Biology','English','Chemistry','Philosophy']}
depts = {'dept_id':[55001,55002,55003,55004,55005], 'dept_name':['Science', 'Science', 'Science', 'Languages', 'Math'], 'subject':['Physics','Chemistry','Biology','English', 'Algebra'], 'difficulty':[3,3,3,1,3]}

emp_df = pd.DataFrame(emps)
dept_df = pd.DataFrame(depts)

append_df = emp_df.append(dept_df)
concat_df = pd.concat([emp_df, dept_df])

print('append_df.equals(concat_df): {}'.format(append_df.equals(concat_df)))

append_df
```

```python
append_df.reset_index()
```

```python
append_df2 = emp_df.append(dept_df, ignore_index=True)
append_df2
```

<a id=MERG><a/>
### Merging <br />
 - [Inner](#in) <br />
 - [Outer](#out) <br />
 - [Left](#lt) <br />
 - [Right](#rt) <br />

```python
emps = {'emp_id':[11001,11002,11003,11004,11005], 'emp_name':['Bob','Mary','Peter', 'Linda','Sam'], 'subject':['Physics','Biology','English','Chemistry','Philosophy']}
depts = {'dept_id':[55001,55002,55003,55004,55005], 'dept_name':['Science', 'Science', 'Science', 'Languages', 'Math'], 'subject':['Physics','Chemistry','Biology','English', 'Algebra'], 'difficulty':[3,3,3,1,3]}

emp_df = pd.DataFrame(emps)
dept_df = pd.DataFrame(depts)
```

<a id=in></a>
#### Inner(Default)
Both DFs should have data, aka `Intersection`

`(1234)+(3456) = ((34))`

```python
print(pd.merge(left=emp_df, right=dept_df, on='subject'))
# emp_df.merge(right=dept_df, on='subject')
```

<a id=out></a>
#### Outer
AKA `Union`

`(1234)+(3456) = (12(34)56)`

```python
print(pd.merge(left=emp_df, right=dept_df, on='subject', how='outer'))
```

<a id=lt></a>
#### Left
(1234)+(3456) = (1234)()`

```python
print(pd.merge(left=emp_df, right=dept_df, on='subject', how='left'))
```

<a id=rt></a>
#### Right
`(1234)+(3456) = ()(3456)`

```python
print(pd.merge(left=emp_df, right=dept_df, on='subject', how='right'))
```

## Multi Index DataFrames

```python

hotels = {
    'hotel': ['Bread Box', 'Bread Box', 'Bread Box', 'Go China', 'Go China', 'Go China', 'Kebabs', 'Kebabs', 'Kebabs'],
    'food_court_id': [1,2,3,1,2,3,1,2,3],
    'tables_served': [12,34,44,76,23,45,98,45,23],
    'items_sold': [122,234,546,348,655,655,654,354,324]
}

hotels_df = pd.DataFrame(hotels)
display(hotels_df)

display(hotels_df.set_index('hotel'))

hotels_df.set_index(['hotel','food_court_id'], inplace=True)
display(hotels_df)

hotels_df.index
```

```python
display(hotels_df.loc['Kebabs'])

display(hotels_df.loc['Kebabs'].loc[1])
```

```python
display(hotels_df)
hotels_df.reset_index(inplace=True)
display(hotels_df)

hotels_df.set_index(['food_court_id', 'hotel'], inplace=True)
display(hotels_df)

display(hotels_df.loc[1])

display(hotels_df.loc[1].loc['Kebabs'])

hotels_df.index
```

```python
display(hotels_df)

display(hotels_df.xs(1, level='food_court_id'))

display(hotels_df.xs('Kebabs', level='hotel'))
```
